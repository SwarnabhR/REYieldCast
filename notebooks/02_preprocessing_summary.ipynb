```python
import pandas as pd
import matplotlib.pyplot as plt

solar = pd.read_csv("../data/processed/solar_gen_2020_mw.csv", index_col=0, parse_dates=True)
wind = pd.read_csv("../data/processed/wind_gen_2020_mw.csv", index_col=0, parse_dates=True)
```

```python
print("Solar shape: ", solar.shape)
print("Wind shape: ", wind.shape)

solar.head()
wind.head()
```

```python
print(solar.isna().sum().sum(), "NaNs in Solar")
print(wind.isna().sum().sum(), "NaNs in Wind")
```

```python
solar[[solar.columns[0], solar.columns[1]]].plot(figsize=(15,4), title="Solar Plant Generation (MW)")
wind[[wind.columns[0], wind.columns[1]]].plot(figsize=(15,4), title="Wind Plant Generation(MW)")
```

```python
solar.sum(axis=1).plot(title="Total Solar Generation")
```

```python
wind.sum(axis=1).plot(title="Total Wind Generation")
```

```python
solar.resample("M").sum().plot(title="Monthly Solar Generation (MW)")
wind.resample("M").sum().plot(title="Monthly Wind Generation (MW)")
```

```python
solar.sample(3, axis=1).plot(title="Sample Soalr PLants - MW over Time")
```

```python
solar.corr().mean().mean()
```
